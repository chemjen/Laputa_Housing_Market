```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 80)
from statsmodels.graphics.gofplots import qqplot
from sklearn.preprocessing import power_transform
from scipy import stats
from  matplotlib.ticker import PercentFormatter
```

```python
df_train = pd.read_csv('train_cleaned.csv')
df_test = pd.read_csv('test_cleaned.csv')
```

```python
target = np.log1p(df_train.SalePrice)
```

```python
df_train.drop('SalePrice', axis = 1, inplace = True)
```

```python
features = df_train.iloc[:,1:]
```

```python
lasso = Lasso(normalize = True)

params = {'alpha':(0.00001,0.0001,0.0002,0.0005,0.001,0.01,0.1,0.5,1,2,10,20,50,100,200,500), 'fit_intercept':[True,False]}
grid_search_lasso = GridSearchCV(estimator=lasso, param_grid=params )
grid_search_lasso.fit(features,target)
```

```python
grid_search_lasso.best_params_
```

```python
grid_search_lasso.best_score_
```
